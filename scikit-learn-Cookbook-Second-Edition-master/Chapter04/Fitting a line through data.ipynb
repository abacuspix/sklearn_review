```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_absolute_error, mean_squared_error
from scipy.stats import probplot

# Load the California housing dataset
california = datasets.fetch_california_housing()

# Initialize the Linear Regression model
lr = LinearRegression()

# Fit the model to the data
lr.fit(california.data, california.target)

# Make predictions
predictions = lr.predict(california.data)

# Calculate and print MAE and MSE
mae = mean_absolute_error(california.target, predictions)
mse = mean_squared_error(california.target, predictions)
print('Mean Absolute Error (MAE):', mae)
print('Mean Squared Error (MSE):', mse)

# Calculate residuals
residuals = california.target - predictions

# Plotting the histogram of residuals
pd.Series(residuals).hist(bins=50)
plt.title('Histogram of Residuals')
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.show()

# Plotting the Q-Q plot
f = plt.figure(figsize=(7, 5))
ax = f.add_subplot(111)
tuple_out = probplot(residuals, plot=ax)
plt.title('Q-Q Plot of Residuals')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from scipy.stats import probplot

# Enable inline plotting in Jupyter notebooks
%matplotlib inline    

# Load the California housing dataset
california = datasets.fetch_california_housing()

# Initialize the Linear Regression model
lr = LinearRegression()

# Fit the model to the data
lr.fit(california.data, california.target)

# Make predictions
predictions = lr.predict(california.data)

# Calculate and print MAE and MSE
mae = mean_absolute_error(california.target, predictions)
mse = mean_squared_error(california.target, predictions)
print('Mean Absolute Error (MAE):', mae)
print('Mean Squared Error (MSE):', mse)

# Calculate residuals
residuals = california.target - predictions

# Plotting the histogram of residuals
pd.Series(residuals).hist(bins=50)
plt.title('Histogram of Residuals')
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.show()

# Plotting the Q-Q plot
f = plt.figure(figsize=(7, 5))
ax = f.add_subplot(111)
tuple_out = probplot(residuals, plot=ax)
plt.title('Q-Q Plot of Residuals')
plt.show()

# Normalize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(california.data)

# Initialize the Linear Regression model with normalized data
lr2 = LinearRegression()

# Fit the model to the normalized data
lr2.fit(scaled_data, california.target)

# Make predictions with the normalized data
predictions_normalized = lr2.predict(scaled_data)

# Calculate and print MAE and MSE for the normalized data
mae_normalized = mean_absolute_error(california.target, predictions_normalized)
mse_normalized = mean_squared_error(california.target, predictions_normalized)
print('Mean Absolute Error (MAE) with Normalized Data:', mae_normalized)
print('Mean Squared Error (MSE) with Normalized Data:', mse_normalized)

# Calculate residuals for the normalized data
residuals_normalized = california.target - predictions_normalized

# Plotting the histogram of residuals for normalized data
pd.Series(residuals_normalized).hist(bins=50)
plt.title('Histogram of Residuals (Normalized Data)')
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.show()

# Plotting the Q-Q plot for normalized data
f = plt.figure(figsize=(7, 5))
ax = f.add_subplot(111)
tuple_out = probplot(residuals_normalized, plot=ax)
plt.title('Q-Q Plot of Residuals (Normalized Data)')
plt.show()
```
