```python
from sklearn import datasets
from sklearn import cluster

blobs, ground_truth = datasets.make_blobs(1000, centers=3,cluster_std=1.75)
```

```python
from sklearn import datasets
from sklearn import cluster

%matplotlib inline
import matplotlib.pyplot as plt
 
f, ax = plt.subplots(figsize=(7, 5))
colors = ['r', 'g', 'b']
for i in range(3):
     p = blobs[ground_truth == i]
     ax.scatter(p[:,0], p[:,1], c=colors[i],
     label="Cluster {}".format(i))
ax.set_title("Cluster With Ground Truth")
ax.legend()
```

```python
kmeans = cluster.KMeans(n_clusters=3)
kmeans.fit(blobs)
```

```python
kmeans.cluster_centers_
```

```python
f, ax = plt.subplots(figsize=(7, 5))
colors = ['r', 'g', 'b']
for i in range(3): 
     p = blobs[ground_truth == i]
     ax.scatter(p[:,0], p[:,1], c=colors[i], label="Cluster {}".format(i))
```

```python
# Evaluate clustering performance for each class
for i in range(3):
    accuracy = (kmeans.labels_ == ground_truth)[ground_truth == i].astype(int).mean()
    print(f"Accuracy for class {i}: {accuracy:.2f}")
```
