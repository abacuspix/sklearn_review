```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()     #Instantiate tree class
dtc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = dtc.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
import numpy as np
from sklearn import tree
from sklearn.datasets import load_iris
import pydotplus
from IPython.display import Image
from io import StringIO  # Use StringIO from the io module

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Train a DecisionTreeClassifier
dtc = tree.DecisionTreeClassifier()
dtc.fit(X, y)

# Export the decision tree to DOT format
dot_iris = StringIO()
tree.export_graphviz(dtc, out_file=dot_iris, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True, special_characters=True)

# Create a graph from the DOT data
graph = pydotplus.graph_from_dot_data(dot_iris.getvalue())

# Display the graph as an image
Image(graph.create_png())

```
